```python
#Linear Regression Focus

#(a) Descriptive Statistics
import pandas as pd

data = {
    "Hours_Studied": [2,3,5,4,6,7,8,5,3,4],
    "Exam_Score": [50,55,65,60,70,75,80,68,58,62]
}
df = pd.DataFrame(data)

print("Mean:\n", df.mean())
print("Median:\n", df.median())
print("Mode:\n", df.mode())
print("Standard Deviation:\n", df.std())
```

(b) Histogram

```python
import matplotlib.pyplot as plt

df.hist()
plt.show()
```

(c) Correlation

```python
print(df.corr())
```

(d1) Train Linear Regression Model

```python
from sklearn.linear_model import LinearRegression

X = df[["Hours_Studied"]]
y = df["Exam_Score"]

model = LinearRegression()
model.fit(X, y)
```

(d2) Predict for 6.5 Hours

```python
prediction = model.predict([[6.5]])
print("Predicted Score:", prediction[0])
```

(d3) R² and MAE

```python
from sklearn.metrics import r2_score, mean_absolute_error

y_pred = model.predict(X)

print("R2 Score:", r2_score(y, y_pred))
print("MAE:", mean_absolute_error(y, y_pred))
```

2. K-MEANS CLUSTERING (Each Task Separate)

(a) Descriptive Statistics

```python
data = {
    "Age": [22,25,47,52,46,23,35,37,50,28],
    "Monthly_Spend": [250,300,500,550,480,270,400,420,530,320]
}
df = pd.DataFrame(data)

print(df.mean())
print(df.median())
print(df.std())
```

(b) Violin Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.violinplot(data=df)
plt.show()
```

(c) Correlation

```python
print(df.corr())
```

(d1) Apply K-Means (K=2)

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2, random_state=0)
df["Cluster"] = kmeans.fit_predict(df)

print(df)
```

(d2) Scatter Plot

```python
plt.scatter(df["Age"], df["Monthly_Spend"], c=df["Cluster"])
plt.xlabel("Age")
plt.ylabel("Monthly Spend")
plt.show()
```

(d3) Centroids

```python
print("Centroids:\n", kmeans.cluster_centers_)
```

(d4) WCSS

```python
print("WCSS:", kmeans.inertia_)
```

3. HIERARCHICAL CLUSTERING

(a) Descriptive Statistics

```python
data = {
    "Experience_Years": [1,2,5,6,7,3,4,8,2,5],
    "Salary": [30,35,50,55,60,40,45,65,38,52]
}
df = pd.DataFrame(data)

print(df.mean())
print(df.median())
print(df.std())
```

(b) Box Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(y=df["Salary"])
plt.show()
```

(c) Correlation

```python
print(df.corr())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering

# Dataset
data = {
    "Experience_Years": [1,2,5,6,7,3,4,8,2,5],
    "Salary": [30,35,50,55,60,40,45,65,38,52]
}
df = pd.DataFrame(data)

# (d1) Dendrogram
linked = linkage(df, method='ward')

plt.figure(figsize=(8,5))
dendrogram(linked)
plt.title("Dendrogram")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()

# (d2) Agglomerative Clustering
model = AgglomerativeClustering(n_clusters=2)
df["Cluster"] = model.fit_predict(df)

print("Clustered Data:\n", df)

# (d3) Centroids
centroids = df.groupby("Cluster").mean()
print("\nCluster Centroids:\n", centroids)
```
